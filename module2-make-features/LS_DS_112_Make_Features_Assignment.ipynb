```python
# Looks good for the most part, except for that first line and last two, so I'll load the data
# into a DataFrame and snip off the parts I don't need
df = pd.read_csv("LoanStats_2018Q4.csv", skiprows=1, skipfooter=2)
print(df.shape)
df.head()
```

```python
# Alright, looks like I'm not dealing with any structure problems, so lets start cleaning up the data
# First things first, lets see if there are any columns that I can drop right away because they're filled
# with NaNs
print(df.shape)
df.isnull().sum().sort_values()
```

```python
# Alright, so it looks like we've got four rows at the bottom whose number of NaNs is equal to the number of rows
# Since there's no data in them, we can't really use them for anything or infer what they're for,
# so we can just drop them
df = df.drop(["id", "desc", "member_id", "url"], axis=1)
```

```python
# Let's check the number of null values again
print(df.shape)
df.isnull().sum().sort_values()
```

```python
# Okay, that looks good, next we want to start cleaning up some of the types of the columns
# First lets increase the number of columns pandas will allow us to see
pd.set_option("display.max_rows", 150)

df.dtypes
```

```python
# Lets fix up a few columns; why is int_rate an object?
df.int_rate.head()
```

```python
# Those are obviously numbers being turned into strings because they're percentages
# There are also some spaces in front that need to be removed
# Let's write a function to fix this
def int_rate_to_float(int_rate_in):
  if isinstance(int_rate_in, float):  # catch any NaNs just in case
    return int_rate_in
  else:
    return float(int_rate_in.strip().strip("%"))

# I'll just overwrite the current int_rate column with the new data
df.int_rate = df.int_rate.apply(int_rate_to_float)
df.int_rate.head(10)
```

```python
# Lets also check to see if there any NaN values in our floats
df.int_rate[df.int_rate.isnull()]
```

```python
# Looks good!

# Next, lets clean up the employment titles
# Let's start by getting an idea of how many of each kind there are
df.emp_title.value_counts(dropna=False).head(20)
```

```python
# I included NaNs in my list just in case, seems like there are a lot
# also there are other issues, like how there's one labelled "Manager" and another called "manager"
# Obviously, those are the same thing, so let's start by standardizing the titles
def standardize_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"

# And now, to apply it to the column
```

```python
df.emp_title = [standardize_title(title) for title in df.emp_title]
```

```python
df.emp_title.value_counts(dropna=False).head(20)
```

```python
# With the updated list, it seems like it's fixed a lot of problems
# A lot of occupations shifted in the list, too
# I'm also going to replace non-top-20 jobs with "Other"

# Let's start by generating a list of the top 20 jobs
toptwenty = []
for num in range(0,20):
  toptwenty.append(df.emp_title.value_counts().index[num])

# print(toptwenty)

# Alright, the list is set up, letch check every job type and if it's not in this list, rename it to "Other"
def cull_unimportant_titles(title):
  for check in toptwenty:
    if title.startswith(check):
      return check
  return "Other"

df.emp_title = [cull_unimportant_titles(title) for title in df.emp_title]
df.emp_title.head(10)
```

```python
# I think it worked, let's check the stats
df.emp_title.value_counts()
```

```python
# Okay, so it looks like there are a fair amount of NaNs
# I think the missing values in the _date columns means "not applicable" rather than missing data
# So I'm not going to try and replace them, but I will get all of the applicable dates set to a proper format
# This means that functions on columns ending in "_date" need to account for that, but whatever

# I'm gonna do a try/except to catch the NaNs and replace them with NaTs because
# to_datetime is not catching them and being angry at me
def to_datetime_catch_NaNs(dateIn):
  try:
    return pd.to_datetime(dateIn, infer_datetime_format=True)
  except:
    return pd.NaT

df[date_columns] = df[date_columns].apply(to_datetime_catch_NaNs)
print(df[date_columns].dtypes)
df[date_columns].head(10)
```
